```python
import pandas as pd
import sklearn as tree
df = pd.read_csv('titanic.csv')
input = df.drop(['Survived','Name'],axis='columns')
input = input.dropna(axis = 0)
input = input.reset_index()
target = df['Survived']
input.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
input['PassengerId_n'] = le.fit_transform(input['PassengerId'])
input['Sex_n'] = le.fit_transform(input['Sex'])
input['Fare_n'] = le.fit_transform(input['Fare'])
input['Pclass_n'] = le.fit_transform(input['Pclass'])
input['Embarked_n'] = le.fit_transform(input['Embarked'])
input.head()
```
